```python
from matplotlib import pyplot as plt;
from pylab import genfromtxt;

plt.rcParams["figure.figsize"] = [10.00, 5.0]
plt.rcParams["figure.autolayout"] = True

firstfiledata = genfromtxt("INK7D.txt",skip_header=4);
bfiledata = genfromtxt("INK750.txt",skip_header=4);
cfiledata = genfromtxt("INK752.txt",skip_header=4);
secondfiledata = genfromtxt("INK754.txt",skip_header=4);
thirdfiledata = genfromtxt("INK756.txt",skip_header=4);
forthfiledata = genfromtxt("INK758.txt",skip_header=4);

plt.plot(firstfiledata[:, 1],(bfiledata[:, 2]-firstfiledata[:, 2])*10247933*100, label="E.Q.E for 0,44 μW ");
plt.plot(firstfiledata[:, 1],(cfiledata[:, 2]-firstfiledata[:, 2])*3067408*100, label="E.Q.E for 1,47 μW ");
plt.plot(firstfiledata[:, 1],(secondfiledata[:, 2]-firstfiledata[:, 2])*1440604*100, label="E.Q.E for 3,13 μW ");
plt.plot(secondfiledata[:, 1],(thirdfiledata[:, 2]-firstfiledata[:, 2])*898225.3*100, label="E.Q.E for 5,02 μW ");
plt.plot(thirdfiledata[:, 1], (forthfiledata[:, 2]-firstfiledata[:, 2])*572219.7*100, label="E.Q.E for 7,88 μW ");
plt.legend();
plt.title("Au/MWCNTs lq/Si3N4/n-Si/Au E.Q.E under UV illumination")
plt.xlabel('Bias(V)',fontweight ="bold",size=20)
plt.ylabel('E.Q.E (%)',fontweight ="bold",size=20)
plt.ylim(0,50)
plt.show();
```
